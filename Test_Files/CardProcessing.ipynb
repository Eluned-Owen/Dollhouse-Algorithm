```python
import random
model_list = ["Risk Model", "Unpredictability Model", "Concealment Model", "Obsession Model"]
picked = random.randrange(0, 3)

model_picked = []
model_picked.append(model_list[picked])
```

```python
import pandas
cards = pandas.read_csv("behavioural_surplus.csv")
cards
```

```python
collected_cards = [1, 3]
score_tally = []
for i in collected_cards:
    card_finder = cards.iloc[i]
    score = card_finder[model_picked].item()

    cards.loc[i, ' Player One'] += 1
    if cards.loc[i, ' Player One'] == 3:
        print("You can use", card_finder['Card'], "only once more")
    if cards.loc[i, ' Player One'] >= 4:
        print("You've used", card_finder['Card'], "too much")
        break
    else:
        score_tally.append(score)

cards
```

```python
turn_score = 0

for item in score_tally:
    turn_score += item

print(turn_score)
```
