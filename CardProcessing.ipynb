<h1>Card Processing</h1>
<p>This program reads the serial port from COM4 (Arduino Leonardo running the NFC-Input code) and saves the data a a string. The data exported from the ardunio are the behavioural surplus cards' NFC reader tags, written in the app "NFC Tools" for the IPhone.</p>

```python
#pip install pyserial
```

```python
#Importing a module for reading serial communications
import serial

#Connecting to COM4 where the Arduino's USB is connected at with a baud rate of 9600 (same as the "NFC-Input" program)
ser = serial.Serial('COM4', 9600, timeout=1)
print(f"Connected to {ser.name}")
```

```python
#If there's bytes waiting at the input buffer, read them and save them to a card variable
if ser.in_waiting > 0:
    available = ser.read(ser.in_waiting)
    print(f"Available data: {available}")
    card = available
```

```python
#As the card variable is a byte, convert cad to a string using utf-8
#print(type(card))
new_card = card.decode("utf-8")
print(new_card)
```
